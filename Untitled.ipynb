Топ-10 самых продаваемых товаров по количеству и выручке в одном запросе

```python
df.groupby('StockCode')['Quantity'].sum().sort_values(ascending = False).head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Предполагаем, что df уже загружен и содержит данные
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Топ-10 по количеству
top_qty = (
    df.groupby(['StockCode', 'Description'])['Quantity']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# Топ-10 по выручке
top_rev = (
    df.groupby(['StockCode', 'Description'])['Revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# Построение графиков
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# График по количеству
top_qty.sort_values().plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Топ-10 товаров по количеству продаж')
axes[0].set_xlabel('Количество')
axes[0].set_ylabel('Товар')

# График по выручке
top_rev.sort_values().plot(kind='barh', ax=axes[1], color='orange')
axes[1].set_title('Топ-10 товаров по выручке')
axes[1].set_xlabel('Выручка')
axes[1].set_ylabel('Товар')

plt.tight_layout()
plt.show()
```
